```python
from harmony.schemas.requests.text import Instrument, Question
```

```python
instrument_english_orig = example_instruments['GHQ 12 English']
```

```python
english_questions_orig = """been able to concentrate on whatever you’re doing?
lost much sleep over worry?
felt that you are playing a useful part in things?
felt capable of making decisions about things?
felt constantly under strain?
felt you couldn’t overcome your difficulties?
been able to enjoy your normal day-to-day activities?
been able to face up to your problems?
been feeling unhappy and depressed?
been losing confidence in yourself?
been thinking of yourself as a worthless person?
been feeling reasonably happy, all things considered?""".split("\n")
```

```python
instrument_ghq12_shona_google_translated = Instrument(questions=[Question(question_text=q, question_no = str(i)) for i, q in enumerate("""wakakwanisa kuisa pfungwa pane zvese zvauri kuita?
kutadza kurara pamusana pokuzvidya mwoyo?
wakanzwa kuti uri kuita rutivi runobatsira muzvinhu?
wakanzwa uchikwanisa kuita zvisarudzo pamusoro pezvinhu?
ainzwa achinetseka nguva dzose?
wakanzwa kuti haugone kukunda matambudziko ako?
wakakwanisa kunakidzwa nezvaunoita zvezuva nezuva?
wakakwanisa kutarisana nematambudziko ako?
wanga usingafari uye wakaora mwoyo?
wanga uchirasa kuzvivimba?
wanga uchizviona semunhu asina maturo?
wanga uchinzwa kufara zvine mutsindo, uchifunga zvese?""".split("\n"))])
```

```python
def get_shona_embedding(texts: list):
  shona_embeddings = model(texts)
  return np.asarray([np.mean(token_embeddings[0], axis=0) for token_embeddings in shona_embeddings])
```

```python
# import os
# os.environ["HARMONY_NO_PARSING"] = "True"
# os.environ["HARMONY_NO_MATCHING"] = "True"
```

```python
from harmony import match_instruments_with_function
```

```python
shona_instrument = Instrument(questions = [Question(question_text=q, question_no = str(i)) for i, q in enumerate(questions.split("\n"))])
```

```python
all_questions, similarity, query_similarity, new_vectors_dict = match_instruments_with_function([instrument_ghq12_shona_google_translated, shona_instrument], None, get_shona_embedding)
```

```python
threshold = 0.5
```

```python
pair_to_similarity = dict()
for q1 in range(len(instrument_ghq12_shona_google_translated.questions)):
    for q2 in range(len(instrument_ghq12_shona_google_translated.questions), similarity.shape[1]):
        s = similarity[q1][q2]
        pair_to_similarity[(q1, q2)] = s
```

```python
import operator
```

```python
seen = set()
for (q1, q2), score in sorted(pair_to_similarity.items(), key=operator.itemgetter(1), reverse=True):
    if q1 in seen or q2 in seen:
      continue
    print (f"{score:.0%}\tShona GHQ-12 \"{all_questions[q1].question_text}\" matched to Shona symptom questionnaire: \"{all_questions[q2].question_text}\"")
    print ("\t\ttranslation of GHQ-12:\t", english_questions_orig[int(all_questions[q1].question_no)])
    print ("\t\ttranslation of SSQ:\t", official_translation[int(all_questions[q2].question_no)])
    seen.add(q1)
    seen.add(q2)
```
